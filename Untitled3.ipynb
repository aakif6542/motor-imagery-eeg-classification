```python
import mne
import numpy as np
import os

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.callbacks import EarlyStopping
```

```python
data_path = r"E:\Downloads\BCICIV_2a_gdf"   

files = sorted([f for f in os.listdir(data_path) if f.endswith("T.gdf")])
all_epochs = []

for file in files:
    print("Loading:", file)
    
    raw = mne.io.read_raw_gdf(os.path.join(data_path, file), preload=True)
    
    events, event_id = mne.events_from_annotations(raw)
    
    epochs = mne.Epochs(
        raw,
        events,
        event_id={'left': 7, 'right': 8},
        tmin=0,
        tmax=4,
        preload=True,
        event_repeated='drop',
        baseline=None
    )
    
    all_epochs.append(epochs)

print("Loaded subjects:", len(all_epochs))
```

```python
epochs = mne.concatenate_epochs(all_epochs)

X = epochs.get_data()
y = epochs.events[:, -1]

print("Combined shape:", X.shape)
print("Labels:", np.unique(y))
```

```python
y = (y == 8).astype(int)

print("Final labels:", np.unique(y))
```

```python
train_files = files[:6]   # A01–A06
test_files = files[6:]    # A07–A09

train_epochs = []
test_epochs = []

for file in train_files:
    raw = mne.io.read_raw_gdf(os.path.join(data_path, file), preload=True)
    events, _ = mne.events_from_annotations(raw)

    ep = mne.Epochs(
        raw, events,
        event_id={'left': 7, 'right': 8},
        tmin=0, tmax=4,
        preload=True,
        event_repeated='drop',
        baseline=None
    )
    train_epochs.append(ep)

for file in test_files:
    raw = mne.io.read_raw_gdf(os.path.join(data_path, file), preload=True)
    events, _ = mne.events_from_annotations(raw)

    ep = mne.Epochs(
        raw, events,
        event_id={'left': 7, 'right': 8},
        tmin=0, tmax=4,
        preload=True,
        event_repeated='drop',
        baseline=None
    )
    test_epochs.append(ep)

# Combine
train_epochs = mne.concatenate_epochs(train_epochs)
test_epochs = mne.concatenate_epochs(test_epochs)

# Extract
X_train = train_epochs.get_data()
y_train = (train_epochs.events[:, -1] == 8).astype(int)

X_test = test_epochs.get_data()
y_test = (test_epochs.events[:, -1] == 8).astype(int)

print(X_train.shape, X_test.shape)
```

```python
mean = np.mean(X_train, axis=(0,2), keepdims=True)
std = np.std(X_train, axis=(0,2), keepdims=True)

X_train = (X_train - mean) / std
X_test = (X_test - mean) / std
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]

print("CNN shape:", X_train.shape)
```

```python
model = Sequential([
    Input(shape=X_train.shape[1:]),

    Conv2D(16, (3, 10), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((1, 3)),

    Conv2D(32, (1, 10), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((1, 3)),

    Flatten(),

    Dense(128, activation='relu'),
    Dropout(0.5),

    Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
early_stop = EarlyStopping(patience=5, restore_best_weights=True)
```

```python
history = model.fit(
    X_train, y_train,
    epochs=45,
    batch_size=32,
    validation_data=(X_test, y_test)
)
```

```python
loss, acc = model.evaluate(X_test, y_test)

print("Final CNN Accuracy:", acc)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['Train', 'Validation'])
plt.title("Model Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = (model.predict(X_test) > 0.5).astype(int)

cm = confusion_matrix(y_test, y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
from mne.decoding import CSP
from sklearn.svm import SVC

# Remove last dimension for CSP
X_train_csp_input = X_train.squeeze()
X_test_csp_input = X_test.squeeze()

csp = CSP(n_components=4, log=True)

X_train_csp = csp.fit_transform(X_train_csp_input, y_train)
X_test_csp = csp.transform(X_test_csp_input)

svm = SVC(kernel='linear')
svm.fit(X_train_csp, y_train)

y_pred_svm = svm.predict(X_test_csp)

svm_acc = accuracy_score(y_test, y_pred_svm)

print("CSP + SVM Accuracy:", svm_acc)
```

```python
from tensorflow.keras.models import Model

# Build model graph
model.predict(X_train[:1])

feature_extractor = Model(
    inputs=model.inputs,
    outputs=model.layers[-3].output
)

X_train_cnn_feat = feature_extractor.predict(X_train)
X_test_cnn_feat = feature_extractor.predict(X_test)
```

```python
# Combine features
X_train_combined = np.concatenate([X_train_csp, X_train_cnn_feat], axis=1)
X_test_combined = np.concatenate([X_test_csp, X_test_cnn_feat], axis=1)

# Train classifier
hybrid_clf = SVC(kernel='linear')
hybrid_clf.fit(X_train_combined, y_train)

y_pred_hybrid = hybrid_clf.predict(X_test_combined)

hybrid_acc = accuracy_score(y_test, y_pred_hybrid)

print("Hybrid Model Accuracy:", hybrid_acc)
```
